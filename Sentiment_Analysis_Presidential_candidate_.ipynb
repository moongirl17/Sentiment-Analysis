# **Sentiment analysis**

```python
# Import library TextBlob untuk analisis sentimen
!pip install textblob
from textblob import TextBlob

# Membuat fungsi untuk mendapatkan subjectivity dari teks
# Menggunakan TextBlob untuk mendapatkan nilai subjectivity dari teks
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
# Membuat fungsi untuk mendapatkan polarity dari teks
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
# Membuat fungsi untuk menentukan analisis sentimen berdasarkan skor polarity
def getAnalysis(score):
    if score < 0:
        return 'negative'  # Jika skor polarity negatif, maka sentimen negatif
    elif score==0:
        return 'neutral'   # Jika skor polarity nol, maka sentimen netral
    else:
        return 'positive'  # Jika skor polarity positif, maka sentimen positif
```

Penjelasan Singkat:

**getSubjectivity:** Fungsi ini digunakan untuk mendapatkan tingkat subjektivitas
dari teks, yaitu seberapa subjektif atau objektif teks tersebut. Nilai berkisar antara 0 (objektif) hingga 1 (subjektif).

**getPolarity:** Fungsi ini digunakan untuk mendapatkan tingkat polaritas dari teks, yaitu seberapa positif, netral, atau negatif teks tersebut. Nilai berkisar antara -1 (negatif) hingga 1 (positif).

**getAnalysis:** Fungsi ini mengkategorikan teks berdasarkan nilai polaritasnya menjadi 'negative', 'neutral', atau 'positive'.

```python

# DataFrame yang telah dibentuk sebelumnya
ganjar_tweets = Data_Mixed.query('(candidat == "Ganjar")')[['full_text']]

# Reset index dan proses analisis
ganjar_tweets.reset_index(inplace=True, drop=True)
# Membersihkan teks dan melakukan analisis sentimen
ganjar_tweets['cleaned_text'] = ganjar_tweets['full_text'].apply(clean_text)
ganjar_tweets['subjectivity'] = ganjar_tweets['cleaned_text'].apply(getSubjectivity)
ganjar_tweets['polarity'] = ganjar_tweets['cleaned_text'].apply(getPolarity)
ganjar_tweets['analysis'] = ganjar_tweets['polarity'].apply(getAnalysis)

# Initialize sentiment score counters
total_positive_score = 0
total_negative_score = 0
total_neutral_score = 0

# Menghitung jumlah kemunculan setiap sentimen (positif, negatif, netral)
sentiment_counts = ganjar['sentiment'].value_counts()

# Aggregate scores
total_score = total_positive_score + total_negative_score + total_neutral_score


# Menghitung skor agregat untuk setiap kategori sentimen
total_positive_score = ganjar[ganjar['sentiment'] == 'positive']['polarity'].sum()
total_negative_score = ganjar[ganjar['sentiment'] == 'negative']['polarity'].sum()
total_neutral_score = ganjar[ganjar['sentiment'] == 'neutral']['polarity'].sum()

# Display sentiment distribution
print("Sentiment Distribution:")
print(sentiment_counts)
print("\nAggregate Scores:")
print("Total Positive Score:", total_positive_score)
print("Total Negative Score:", total_negative_score)
print("Total Neutral Score:", total_neutral_score)

```
